```python
#Checking which all columns have over 60% null values
null_value_columns=df.columns[df.isnull().sum()>=0.6*2107]
null_value_columns 
```

```python
#Deleting those columns above and renaming dataframe as df_cleaned
df=df.drop(columns=['Alley', 'Pool QC', 'Fence', 'Misc Feature'])
```

```python
#Replacing missing values of rest of the columns 
def fill_missing_values(df):
    # Iterate over each column in the DataFrame
    for column in df.columns:
        if df[column].dtype in ['int64', 'float64']:  # Check for numerical columns
            median_value = df[column].median()  # Calculate the median
            df[column].fillna(median_value, inplace=True)  # Fill missing values with median
        elif df[column].dtype == 'object':  # Check for categorical columns
            mode_value = df[column].mode()[0]  # Calculate the mode
            df[column].fillna(mode_value, inplace=True)  # Fill missing values with mode
            
    return df
fill_missing_values(df)   
```

```python
#Replacing outliers using IQR
def replace_outliers(df):
    for column in df.select_dtypes(include='number').columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Replace outliers with the bounds
        df[column] = np.where(df[column] < lower_bound, lower_bound, 
                              np.where(df[column] > upper_bound, upper_bound, df[column]))
    
    return df
```

```python
replace_outliers(df)
replace_outliers(numerical_features)
replace_outliers(categorical_features)
```

```python
#Cleaning the data using datetime:
# Convert 'Mo Sold' and 'Yr Sold' to integers
df['Mo Sold'] = df['Mo Sold'].astype(int)
df['Yr Sold'] = df['Yr Sold'].astype(int)

# Combine 'Mo Sold' and 'Yr Sold' into a single datetime column
df['Date Sold'] = pd.to_datetime(df['Yr Sold'].astype(str) + '-' + df['Mo Sold'].astype(str) + '-01')

# Optionally, drop the original 'Mo Sold' and 'Yr Sold' columns
df.drop(columns=['Mo Sold', 'Yr Sold'], inplace=True)

# Display the cleaned DataFrame
print(df)
```

```python
#check if all values in each column is in same format 

def check_all_columns_format(df):
    result = {}
    
    for col in df.columns:
        result[col] = df[col].apply(type).nunique() == 1  # Check if all types are the same
    
    return result
check_all_columns_format(df)
```

```python
#Correlation matrix for continuous variables:
# Define the continuous variables to analyze
continuous_vars = ['SalePrice', 'Lot Frontage', 'Lot Area', 'Garage Area', 'Date Sold']

# Select the relevant columns from the DataFrame
correlation_data = df[continuous_vars]

# Compute the correlation matrix
correlation_matrix = correlation_data.corr()

# Display the correlation matrix
print(correlation_matrix)
#INSIGHTS:-
#1.) Sales Price has the highest positive correlation with Garage Area with around 0.7064 followed
#distantly by Lot Frontage of 0.4315. 
#2.) Even though the Date Sold has the lowest correlation with Sales Price of the property, it worth noting that in the 
#graph (line chart) above, there is a moderate relation of it with the month sold. 
#3.) Therefore, there could be factors such as property condition, overall location of the houses which
#could cause just a moderate positive correlation of Lot Frontage and Area with the Sales Price. 
#4.) Garage Area is significantly impacting the sales price possibly due to as families are larger, there is 
#possible need of more vehicles which will require Garage Area and higher prices overall. Also, garage Area provides
#additional security for the vehicle which could increase house value and price overall. As, remote working has increased,
#there is a greater possibility of converting the garage into a home office which increases the demand for larger garage area 
#thus, proportionately increasing the price of the property.
```

```python
# Replace categorical values with numerical representations, we have not used this for Street, as it is
#not a nominal attribute. We are just checking for ordinal attributes. 
df['Lot Shape'] = df['Lot Shape'].replace({'Reg': 1, 'IR1': 2, 'IR2': 3, 'IR3': 4})
df['Land Contour'] = df['Land Contour'].replace({'Low': 1, 'Lvl': 2, 'Bnk': 3, 'HLS': 4})

# Select the columns of interest and SalePrice
columns_of_interest = ['Lot Shape', 'Land Contour', 'Overall Qual', 'Overall Cond', 'SalePrice']
correlation_data = df[columns_of_interest]

# Compute the correlation matrix
correlation_matrix = correlation_data.corr()

# Extract correlation with SalePrice
saleprice_correlation = correlation_matrix['SalePrice']

# Display the correlation with SalePrice
print(saleprice_correlation)
#INSIGHTS:-
#1.) There is a very strong positive correlation between overall quality and sales price indicating as the quality
#increases, the property value and price increases. 
#2.) There is a moderately positive correlation between Lot Shape of the property and sales price of the property,
#where as the irregularity of the shape could moderately increase the sales price of the property as they could be located in prime/expensive locations.
#3.) Overall a weak negative correlation between Land Contour, overall condition as (mentioned earlier), in some markets,
#the required condition criteria for the properties may not have exceeded 5.0 or 6.0, so overall condition of the property 
# is not a significant driver of the change in sales price. Also, for the level of the land, there is no clear correlation
#as houses with higher lands then have higher median sales prices whereas level (flat level land) houses having higher
#median price than banked ones (sloped) of higher contours, shows how level of land does not impact the sales price that much indicating
#a weaker correlation. 
#4.) Despite all the analysis made above, one must note many of these categories of these ordinal attributes
# of the properties have several outliers for every category. 
```
